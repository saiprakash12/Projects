```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
print("Coefficient of determination: R^2--> on train data set: {}".format(model.score(X_train,y_train)))
```

```python
print("Coefficient of determination: R^2--> on test data set: {}".format(model.score(X_test,y_test)))
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(model,X,y,cv=5)
```

```python
score.mean()
```

```python
score.min()
```

```python
score.max()
```

```python
#Now, R2 is minimum and the before value is 100 %. It is purely overfitting case as our decision tree splitting the nodes using all features till leaf node
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Hyper Parameters
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]
max_features=['auto','sqrt']
max_depth=[int(x) for x in np.linspace(5,30,num=6)]
min_samples_split=[2,5,10,15,20,100]
min_samples_leaf=[1,2,5,10]
```

```python
random_grid={
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf
}
print(random_grid)
```

```python
#Use random grid to search for best hyper parameters
```

```python
rf=RandomForestRegressor()
```

```python
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=100,cv=5,verbose=2,random_state=42)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
predictions=rf_random.predict(X_test)
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
print('MAE: ',metrics.mean_absolute_error(y_test,predictions))
print('MSE: ',metrics.mean_squared_error(y_test,predictions))
print('RMSE: ',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
print("Coefficient of determination: R^2--> on train data set: {}".format(rf_random.score(X_train,y_train)))
```

```python
print("Coefficient of determination: R^2--> on train data set: {}".format(rf_random.score(X_test,y_test)))
```
